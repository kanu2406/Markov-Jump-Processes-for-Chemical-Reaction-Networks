## **1. Markov Jump Process for Chemical Reaction Networks**

In a chemical reaction network, molecules undergo stochastic transitions between reactants and products according to reaction rates. This behavior can be mathematically described using the principles of Markov jump processes, where the state of the system represents the configuration of chemical species present at any given time. The transitions between states occur randomly and depend on the probabilities associated with individual reactions.

First, consider the following reaction -

$$A + B → C$$

The intuition for the model for the reaction
is that the probability of the reaction occurring in a small time interval (t, t + ∆t] should be proportional to the product of the numbers of
molecules of each of the reactants and to the length of the time interval.

### **Rates for the law of mass action -**

* The stochastic form of the law of mass action says that the rate at
which a reaction occurs should be proportional to the number of distinct
subsets of the molecules present that can form the inputs for the reaction.
Assuming that at time t there are $X_{\text{A}}(t)$ molecules of $A$ and $X_{\text{B}}(t)$
molecules of $B$ in our system, rate of our reaction is given by:

$$λ(X_t) = κ X_{\text{A}}(t) X_{\text{B}}(t)$$

and we express our assumption about the
probability of the reaction occurring by:


$$ P(\text{reaction occurs in }(t, t + \Delta t) | \mathcal{F}_t) \approx \kappa X_A(t)X_B(t)\Delta t  \qquad    (1.1)$$

where $ \mathcal{F}_t $ represents the information about the system available at time $t$, and $\kappa$ is a positive constant representing the reaction rate constant.

* The number of times that the reaction occurs by time $t$ is given by $(N_t)_{t∈ℝ_+}$
the counting process satisfying
$$N_t = Y(∫_0^t λ(X_s)ds)$$
where the $Y$ are independent unit Poisson processes.

Indeed, the time-change representation $Y_{λ}(t) = Y(λt)$ explains the above writting of $(N_t)_{t∈ℝ_+}$. Here, $(N_t)_{t∈ℝ_+}$ is a non-homogeneous poison process as its intensity function depends on the time and the position. You will find more information about non-homogeneous Poisson process below.

* The state of the system $(X_t)_{t ∈ ℝ_+}$ wich is the triple $\begin{pmatrix}
X_A(t) \\
X_B(t) \\
X_C(t) \\
\end{pmatrix}_{t ∈ ℝ_+}$ satisfies


$$X_t = X_0 + N_t \begin{pmatrix}
-1 \\
-1 \\
1 \\
\end{pmatrix}
$$

where $Nt$ is the number of times the reaction has occurred by time $t$, $\eta = \begin{pmatrix}
-1 \\
-1 \\
1 \\
\end{pmatrix}$ is the vector that describes the stochiometry of the reaction, indicating how the number of molecules of each species changes due to the reaction and $X_0$ is the vector giving the numbers of molecules of each of the chemical species in the system at time zero. We will assume that two reactions cannot occur at exactly the same time, so $N$ is a counting process, that is, $N_0 = 0$ and $N$ is constant except for jumps of plus one.




### **Non-Homogenous Poisson Process**

**Definition:** The counting process ${N(t), t \geq 0}$ is said to be a nonhomogeneous Poisson process with intensity function $\lambda(t), t \geq 0$, if:
1. $N(0) = 0$.
2. ${N(t), t \geq 0}$ has independent increments.
3. $P\{N(t + h) - N(t) \geq 2\} = o(h)$.
4. $P\{N(t + h) - N(t) = 1\} = \lambda(t)h + o(h)$.


Now, consider a homogenous Poisson process $Y_\lambda(t)$ with intensity $λ$.

Then by the properties of Poisson Process we know that for small time interval $Δt$ -

$$ P\{ Y_\lambda(t + \Delta t) - Y_\lambda(t) > 0 | \mathcal{F}_t \} = P\{ Y_\lambda(t + \Delta t) - Y_\lambda(t) > 0 \}\\ = 1 - e^{-\lambda\Delta t} \approx \lambda\Delta t. $$

Our reaction rate depends on the current time t and resembles Poisson Process. So, we will take our counting Process $N_t$ as a "Non-homogenous Poisson Process"

Therefore, the number of times that the reaction occurs by time $t$ is given by $(N_t)_{t∈ℝ_+}$
the counting process satisfying
$$N_t = Y(∫_0^t λ(X_s)ds)$$
where the $Y$ is independent unit Poisson processes.

Hence,

$$
\begin{pmatrix}
X_A(t) \\
X_B(t) \\
X_C(t) \\
\end{pmatrix}
\equiv X_t = X_0 +
\begin{pmatrix}
-1 \\
-1 \\
1 \\
\end{pmatrix}
Y\left(\int_0^t \kappa X_A(s)X_B(s) ds\right).
$$


### **Markov Jump Process $(X_t)_{t∈ℝ_+}$**

Given some properties of the underlying Poisson process $N_t$, and that the rate of this process is governed by the current state of $X_t$, $X_t$ itself inherits the Markov property. The state transitions occur only at discrete jump times determined by $N_t$, making $X_t$ a Markov Jump Process. The dynamics of $X_t$ clearly illustrate that its future states depend only on its present state and evolve through jumps at random times, satisfying the conditions of being a Markov Jump Process.


### **Proving the Markov Property**

Let's prove that

$$\mathbb{P}(X_{t_n} = x_n | X_{t_{n-1}} = x_{n-1}, X_{t_{n-2}} = x_{n-2}, ..., X_{t_0} = x_0) = \mathbb{P}(X_{t_n} = x_n | X_{t_{n-1}} = x_{n-1})$$

First, we know the state transition is given by
$$ X_{t_n} = X_{t_{n-1}} + (N_{t_n} - N_{t_n})η$$

where $\eta$ is the vector that describes the stochiometry of the reaction.

As $N_t$ is a non homogenous Poisson process, we know that its increments follow the following law  $$N_{t_n} - N_{t_{n-1}} ∼ Poisson(∫_{t_{n-1}}^{t_n} λ(X_s)ds)$$.

Then we can write $(N_{t_n} - N_{t_{n-1}})\eta = X_{t_n} - X_{t_{n-1}}$

Hence, the probability on the left side becomes

$$ \mathbb{P}(X_{t_n} = x_n | X_{t_{n-1}} = x_{n-1}, X_{t_{n-2}} = x_{n-2}, ..., X_{t_0} = x_0) = \mathbb{P}((N_{t_n} - N_{t_{n-1}})\eta = x_n - X_{t_{n-1}}|\mathcal{F}_{t_{n-1}})$$

where $\mathcal{F}_{t_{n-1}}$ is the filtration up to $t_{n-1}$.


Due to continuity in the chemical concentration, we assume that

$\mathbb{P}($reaction occurs in $(t, t+Δt]| \mathcal{F}_t) ≈ λ(X_t) Δt$.

So in our case, $$\mathbb{P}((N_{t_n} - N_{t_{n-1}})\eta|\mathcal{F}_{t_{n-1}}) ≈ λ(X_{t_{n-1}})$$

Hence, Since $(N_{t_n} - N_{t_{n-1}})\eta$ only depends on the state at $t_{n-1}$ through $λ(X_{t_{n-1}})$, the larger filtration simplifies to the immediate past state for determining $(N_{t_n} - N_{t_{n-1}})\eta$. This means:

$$\mathbb{P}((N_{t_n} - N_{t_{n-1}})\eta = x_n - X_{t_{n-1}}|\mathcal{F}_{t_{n-1}}) = \mathbb{P}((N_{t_n} - N_{t_{n-1}})\eta = x_n - X_{t_{n-1}}|X_{t_{n-1}} = x_{n-1})$$

So we proved that
$$\mathbb{P}(X_{t_n} = x_n | X_{t_{n-1}} = x_{n-1}, X_{t_{n-2}} = x_{n-2}, ..., X_{t_0} = x_0) = \mathbb{P}(X_{t_n} = x_n | X_{t_{n-1}} = x_{n-1})$$.

### **Proving the Jump Dynamics**
The system undergoes changes (jumps) in its state only at the times of reaction events, which are counted by $(N_t)_{t∈ℝ_+}$.

## **Simulation of the process**

We will simulate the process inspired from *Gillespie Algorithm*, also known as the stochastic simulation algorithm (SSA). It is a computational method used to simulate the time evolution of chemical reaction systems with inherent randomness.It was developed by Daniel T. Gillespie in 1976.

At its core, the Gillespie algorithm operates by considering each reaction event as a stochastic process. It proceeds in a step-by-step fashion, selecting the next reaction to occur and the time interval until that event happens based on the propensity functions of the reactions and random numbers drawn from appropriate probability distributions.

```python
import numpy as np
import scipy as sc
import scipy.stats as scs
import matplotlib.pyplot as plt
from numpy import zeros, cumsum, arange, append
from numpy.random import poisson, rand, exponential
from matplotlib.pyplot import figure, step, title, legend
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate the intensity function lambda(t)
def intensity_function(X_A, kappa):
    return kappa * X_A

def update_lambda(t, X, times):
    lambda_integral = np.trapz(kappa * X[0] * X[1] * (np.array(times) <= t), times)  # Compute integral up to time t using trapezoidal rule
    return lambda_integral


# Function to simulate the non-homogeneous Poisson process
def simulate_process(X0, kappa, eta, T):
    # Initialize variables
    t = 0
    X = X0
    N = 0  # Number of reaction events

    # Lists to store time, state of the system, and reaction events
    times = [t]
    states = [X]
    events = [N]
    integral_lambda = [0]  # Integral of intensity function up to current time

    # Simulation loop
    while t < T:
        # Calculate intensity function at current time
        lambda_t = intensity_function(X[0], kappa)

        # Generate time until next event using exponential distribution
        dt_event = np.random.exponential(scale=1 / lambda_t)

        # Update time
        t += dt_event

        # Check if simulation time exceeds T
        if t > T:
            break

        # Increment number of reaction events
        N += 1

        # Update state of the system
        X = X + eta

        new_lambda=update_lambda(t, X, times)
        # Compute integral of intensity function up to current time
        integral_lambda.append(new_lambda)

        # Append data to lists
        times.append(t)
        states.append(X)  # Update state with the effect of reaction events
        events.append(N)

    return times, states, events, integral_lambda
```

```python
# Parameters
X0 = np.array([100, 00, 00])  # Initial number of molecules of A, B, and C
kappa = 0.1  # Reaction rate constant
eta = np.array([-1, 2, 1])
T = 200  # Total simulation time

# Simulate the process
times, states, events, integral_lambda = simulate_process(X0, kappa,eta, T)

# Convert results to NumPy arrays for easy plotting
times = np.array(times)
print(max(times), times.size)
states = np.array(states)
events = np.array(events)
integral_lambda = np.array(integral_lambda)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(times, states[:, 1], 'r--',label='B',linewidth=4)  # Plot species B first
plt.plot(times, states[:, 2], label='C')  # Plot species C next
plt.plot(times, states[:, 0],color='yellow',label='A')  # Plot species A last with adjusted linewidth
plt.xlabel('Time')
plt.ylabel('Number of Molecules')
plt.title('Simulation of Chemical Reaction Process')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting N(t), the number of reactions until t
plt.figure(figsize=(10, 6))
plt.step(times, events, where='post', label='N_t')
plt.xlabel('Time')
plt.ylabel('Number of Reaction Events')
plt.title('Number of Reaction Events over Time')
plt.grid(True)
plt.legend()
plt.show()
```

## **2. Theoritical Result**

For our reaction,

\begin{align}
  A &→ 2B + C
\end{align}

If $X(t) = (X_A(t), X_B(t), X_C(t))$ gives the state of the process at time $t$ and $X(0) = (X_A(0), X_B(0), X_C(0))$ denotes number of molecules of $A$, $B$, and $C$ at time $t=0$. Then,

$$ \mathbb{E}[X_A(t)] = X_A(0) e^{-\kappa t} $$
$$ \mathbb{E}[X_B(t)] = -2X_A(0) e^{-\kappa t} + X_B(0) + 2X_A(0) $$
$$ \mathbb{E}[X_C(t)] = -X_A(0) e^{-\kappa t} + X_C(0) + X_A(0) $$



***Proof:***

For the given reaction,by rate of law of mass action, our itensity will be given by-


$$λ(X(t)) = κ X_{\text{A}}(t)$$
The stochastic equation for the model becomes-



$$
\begin{pmatrix}
X_A(t) \\
X_B(t) \\
X_C(t) \\
\end{pmatrix}
\equiv X(t) = X(0) +
\begin{pmatrix}
-1 \\
2 \\
1 \\
\end{pmatrix}
N\left(\int_0^t \kappa X_A(s) ds\right)    ------(**)    
$$


where $N(t)$ is the number of times the reaction has occurred by time $t$ which follows **Non-homogenous Poisson Process** and $X(0)$ is the vector giving the numbers of molecules of each of the chemical species in the system at time zero. We will assume that two reactions cannot occur at exactly the same time, so $N$ is a counting process, that is, $N(0) = 0$ and $N$ is constant except for jumps of plus one. $κ$  is a positive reaction rate constant.

Now, we will use the following proposition to proceed further-

**Proposition:**
Unlike a homogeneous Poisson process where events occur with a constant average rate, in a Non-homogeneous Poisson Process, the rate of occurrence of events, $λ(t)$, can vary over time hence
$$
\mathbb{E}[N_t] = \int_{0}^{t} \lambda(y) \, dy = m(t)
$$
The function $m(t)$ defined by

$$m(t) = \int_{0}^{t} \lambda(y) \, dy$$
is called the ***mean value function*** of the non-homogeneous Poisson process.

However, in our case, the rate at which reactions occur is $λ(X_t) = κ X_{\text{A}}(t)$ and depends on the concentrations of the reactant $A$. In a standard Poisson process $Y(μ)$ with a deterministic rate $μ$,
$E[Y(μ)]=μ$. Extending this to a Poisson process where the rate $μ$ is itself a random variable (here, $μ=\int_0^t λ(X_s)ds$), we still have:
$$E[Y(μ)]=E[μ]$$
Since, $μ$ is random (not deterministic), provided that $μ$ is integrable and the expectation is well-defined.

Now, using the proposition, we can say that
$$\mathbb{E}[N(\int_{0}^{t} \lambda(X_s) \, ds)] = \mathbb{E}[\int_{0}^{t} \lambda(X_s) \, ds]$$

and by denoting $X_t = X(t)$, we have
$$\mathbb{E}[X(t)] = \mathbb{E}[X(0)] + \begin{pmatrix}
-1 \\
2 \\
1 \\
\end{pmatrix} \int_{0}^{t}  \kappa \mathbb{E}[X_{A}(s)] \, ds $$

$$ \mathbb{E}[X(t)] = \begin{pmatrix}
\mathbb{E}[X_A(t)] \\
\mathbb{E}[X_B(t)] \\
\mathbb{E}[X_C(t)] \\
\end{pmatrix} = \begin{pmatrix}
\mathbb{E}[X_A(0)] \\
\mathbb{E}[X_B(0)] \\
\mathbb{E}[X_C(0)] \\
\end{pmatrix} + \int_{0}^{t} \begin{pmatrix}
-\kappa & 0 &0 \\
2\kappa & 0 &0 \\
\kappa & 0 &0\\
\end{pmatrix} \begin{pmatrix}
\mathbb{E}[X_A(s)] \\
\mathbb{E}[X_B(s)] \\
\mathbb{E}[X_C(s)] \\
\end{pmatrix} ds$$

We can write the expectations for $X_A(t)$,$X_B(t)$ and $X_C(t)$ as following -

$$\mathbb{E}[X_A(t)]=X_A(0)+ \int_{0}^{t} (-\kappa)\mathbb{E}[X_A(s)] ds$$
$$\mathbb{E}[X_B(t)]=X_B(0)+ \int_{0}^{t} 2\kappa \mathbb{E}[X_A(s)] ds$$
$$\mathbb{E}[X_C(t)]=X_C(0)+ \int_{0}^{t} \kappa \mathbb{E}[X_A(s)] ds$$



Theoretically, we will use the method of integrating factors. Let's denote the expected values as $x_A(t)$, $x_B(t)$, and $x_C(t)$ respectively.

Given equations:
$$ \frac{d}{dt} x_A(t) = -\kappa x_A(t) $$
$$ \frac{d}{dt} x_B(t) = 2\kappa x_A(t) $$
$$\frac{d}{dt} x_C(t) = \kappa x_A(t) $$

We can rewrite these equations in terms of differentials:
$$ dx_A(t) = -\kappa x_A(t) dt $$
$$dx_B(t) = 2\kappa x_A(t) dt $$
$$ dx_C(t) = \kappa x_A(t) dt $$

Now, let's integrate both sides of each equation:

1. For $x_A(t)$: $\;$ as $ \frac{dx_A(t)}{x_A(t)} = -κ dt$,

$$ \int \frac{1}{x_A(t)} dx_A(t) = \int -\kappa dt $$
$$ \ln|x_A(t)| = -\kappa t + C_1 $$
$$ x_A(t) = C_2 e^{-\kappa t} $$

2. For $x_B(t)$:
$$\int dx_B(t) = \int 2\kappa x_A(t) dt $$
$$ x_B(t) = 2\kappa \int x_A(t) dt + C_3 $$
$$ \text{using } x_A(t) = C_2 e^{-\kappa t} \text{ we get,}$$

$$ x_B(t) = -2C_2 e^{-\kappa t} + C_4 \text{ where } C_4 = C_3 + \text{ another constant } $$

3. For $x_C(t)$:
$$ \int dx_C(t) = \int \kappa x_A(t) dt $$
$$ x_C(t) = \kappa \int x_A(t) dt + C_5 $$
$$ x_C(t) = -C_2 e^{-\kappa t} + C_6 \text{ where } C_6 = C_5 + \text{ another constant }$$

Now, let's apply the initial conditions:

Given $x_A(0) = X_A(0)$, $x_B(0) = X_B(0)$, and $x_C(0) = X_C(0)$:

1. $x_A(0) = C_2 = X_A(0)$
2. $x_B(0) = -2C_2 + C_4 = X_B(0)$
3. $x_C(0) = -C_2 + C_6 = X_C(0)$

Solving these equations for $C_4$ and $C_6$, we find:

1. $C_4 = X_B(0) + 2X_A(0)$
2. $C_6 = X_C(0) + X_A(0)$

Now, substituting these values back into our solutions for $x_B(t)$ and $x_C(t)$, we get:

$$ x_B(t) = -2X_A(0) e^{-\kappa t} + X_B(0) + 2X_A(0) $$
$$ x_C(t) = -X_A(0) e^{-\kappa t} + X_C(0) + X_A(0) $$

So, the solutions for the expected values are:
$$ \mathbb{E}[X_A(t)] = X_A(0) e^{-\kappa t} $$
$$ \mathbb{E}[X_B(t)] = -2X_A(0) e^{-\kappa t} + X_B(0) + 2X_A(0) $$
$$ \mathbb{E}[X_C(t)] = -X_A(0) e^{-\kappa t} + X_C(0) + X_A(0) $$


## **3. Solving Numerically**

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Define the system of differential equations
def rhs(X, t, kappa):
    dXdt = np.zeros_like(X)
    dXdt[0] = -kappa * X[0]  # dX_A/dt = -kappa * X_A
    dXdt[1] = 2 * kappa * X[0]  # dX_B/dt = 2 * kappa * X_A
    dXdt[2] = kappa * X[0]  # dX_C/dt = kappa * X_A
    return dXdt

# Generate random initial values
X0 = [100,0,0] # Replace with your desired method to generate initial values

# Time points for which to solve the equations
t = np.linspace(0, 50, 100)  # Example time span, adjust as needed

# Value of kappa
kappa = 0.1  # Example value, adjust as needed

# Solve the system of equations
sol = odeint(rhs, X0, t, args=(kappa,))

# Extract solution for each variable
X_A = sol[:, 0]
X_B = sol[:, 1]
X_C = sol[:, 2]

# Plot the solutions
plt.plot(t, X_A, label='$X_A(t)$')
plt.plot(t, X_B, label='$X_B(t)$')
plt.plot(t, X_C, label='$X_C(t)$')
plt.xlabel('Time')
plt.ylabel('Expected Value')
plt.legend()
plt.title('Expected Values over Time')
plt.grid(True)
plt.show()
```

## **Numerical Approximation**

To show the validity of $E[X_t]$, we will simulate $M$ $iid$ $(X_t^{(i)})_{i=1,...,M}$ copies of $X_t$. Using the law of large numbers,

$$ \lim_{{M \to \infty}} \frac{1}{M} \sum_{{i=1}}^{M} {X_{t}}^{(i)} = \mathbb{E}[X_t] \quad \text{a.s.} $$


First, we are defining `Sample_Model` that will return M values of molecules of A, B and C at time T. Then, we defined `LLN_Prob_Model` that will return mean and standard deviation of those values. We will later use these values to calculte expectation and confidence interval

```python
def Sample_Model(M, X0, kappa, eta, T):
    sample=[]
    for i in range(M):
        # Simulate the process
        times, states, events, integral_lambda = simulate_process(X0, kappa, eta, T)
        sample.append(states[-1])
    sample_A=[triple[0] for triple in sample]
    sample_B=[triple[1] for triple in sample]
    sample_C=[triple[2] for triple in sample]
    return sample_A,sample_B,sample_C

###############################################################################
#This function returns the Monte Carlo mean estimator and the variance estimator at time t=T of  for M samples
###############################################################################
def LLN_Prob_Model(M, X0, kappa, eta, T):
    sample_A,sample_B,sample_C = Sample_Model(M, X0, kappa, eta, T)
    return np.mean(sample_A),np.std(sample_A),np.mean(sample_B),np.std(sample_B),np.mean(sample_C),np.std(sample_C)
```

Here, we are initializing parameters for simulating the process in order to perform Monte Carlo approximations of expectations.

```python
X0 = np.array([200, 0, 0])  # Initial number of molecules of A, B, and C
kappa = 0.1  # Reaction rate constant
eta = np.array([-1, 2, 1])
T = 50  # Total simulation time
M = 10000  # Number of simulations

# Estimate E[X(t)] using LLN
mean_A,std_A,mean_B,std_B,mean_C,std_C = LLN_Prob_Model(M, X0, kappa, eta, T)
```

### **Theoreitical Vs Estimated Values**

```python
def theoretical_expected_values(X0, kappa, t):
    X_A_0, X_B_0, X_C_0 = X0
    E_X_A = X_A_0 * np.exp(-kappa * t)
    E_X_B = -2 * X_A_0 * np.exp(-kappa * t) + X_B_0 + 2 * X_A_0
    E_X_C = -X_A_0 * np.exp(-kappa * t) + X_C_0 + X_A_0
    return np.array([E_X_A, E_X_B, E_X_C])

theoretical_values = theoretical_expected_values(X0, kappa, T)
estimated_expected_values=[mean_A,mean_B,mean_C]

# Print and plot results
labels = ['X_A', 'X_B', 'X_C']
for i in range(3):
    print(f"Estimated E[{labels[i]}(t)]: {estimated_expected_values[i]}")
    print(f"Theoretical E[{labels[i]}(t)]: {theoretical_values[i]}")
    print(f"Absolute Error for {labels[i]}: {abs(estimated_expected_values[i] - theoretical_values[i])}\n")
```

The absolute errors for all species are relatively small, indicating consistency between the estimated and theoretical values across the species. This consistency suggests that the Monte Carlo simulations are stable and reliable for this system.

### **Confidence Intervals**

### **Extension: Existence of a Steady-State Distribution**

For our theoritical result, we also wanted to find $\pi$ using $\pi$ Q=0 and then we can verify it numerically using ergodic theorem.

$$ \lim_{{t \to \infty}} \frac{1}{t} \int_0^t f(X_t) \, dt  = E[f(Y)] \text{
    a.s.}$$


 where Y follows same law as $\pi$.
We could use Monte Carlo simulations to show that for $t$ big enough, $E[f(X_t)]$ is approximately equal to $E[f(Y)]$ where $Y ∼ \pi$.


In many chemical reaction networks, there exists a steady-state distribution of the chemical species concentrations. This distribution represents the long-term behavior of the system, where the rates of forward and backward reactions balance out, leading to a constant concentration profile over time.

To prove this result, we wanted touse the concept of detailed balance equation, which states that for any closed loop of reactions in the network, the product of the reaction rate constants in the forward direction is equal to the product of the rate constants in the backward direction. This ensures that the system reaches a state where the rates of reactions in each direction are equal, leading to a steady-state distribution.

Mathematically, we could have expressed the steady-state distribution π as the solution to the system of equations:

$$\pi Q=0$$

Where $π$ is a row vector representing the steady-state distribution, and $Q$ is the infinitesimal generator matrix of the Markov jump process representing the chemical reaction network.

Once we have proven the existence of the steady-state distribution theoretically, we can simulate the chemical reaction network numerically to observe and verify the convergence of the system to this steady state.

By comparing the simulated steady-state distribution with the theoretically predicted distribution, we can validate the theoretical result and gain insights into the long-term behavior of the chemical reaction network.

We could not proceed with this result because it was difficult to find the infinitesimal generator matrix $Q$ as our intensity function $\int_{0}^{t} \lambda(X_y) \, dy$ varies with time. So, $Q$ also depends on $X_t$.

This theoretical result and its numerical validation provide a fundamental understanding of the behavior of chemical reaction networks and can serve as a basis for further analysis and applications in fields such as systems biology, pharmacokinetics, and environmental chemistry.
